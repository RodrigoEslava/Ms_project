```python
#oficial
result = list()

classe = list()

for index, row in first_df.iterrows():

            first_column_from_2nd_df = second_df[second_df["genes"].isin([row['Gene_1']])]
            

            second_column_from_2nd_df = second_df[second_df["genes"].isin([row['Gene_2']])]

            third_column = (1)
            
            resulting_tuple = (first_column_from_2nd_df[[str(i) for i in range(1, 51)]].to_string(header = False, index=False), second_column_from_2nd_df[[str(i) for i in range(1, 51)]].to_string(header = False, index=False))

            classe.append(third_column)
           
            result.append(resulting_tuple)


first_df['concat'] = pd.Series(result)
first_df['classe'] = classe
first_df
```
