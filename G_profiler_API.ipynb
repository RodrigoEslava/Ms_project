```python
import pandas as pd
import gprofiler as gp
import json
import numpy as np

# Criar função para converter int64 para int e float64 para float
def converter_tipo(valor):
    if isinstance(valor, np.int64):
        return int(valor)
    elif isinstance(valor, np.float64):
        return float(valor)
    else:
        return valor

# Ler o DataFrame
df = pd.read_csv('caminho/do/arquivo.csv')

# Criar lista para armazenar os resultados
resultados = []

# Iterar sobre cada linha do DataFrame
for i, row in df.iterrows():
    # Realizar a query no GProfiler
    res = gp.profile([converter_tipo(row['Coluna 1']), converter_tipo(row['Coluna 2'])], organism='hsapiens', user_threshold=0.05)
    
    # Verificar se o resultado atende às condições
    resultado = 0
    for r in res:
        if 'source' in r and ('KEGG' in r['source'] or 'BP' in r['source']) and r['intersection_size'] == 2:
            resultado = 1
            break

    # Adicionar o resultado à lista de resultados
    resultados.append(resultado)

# Adicionar a lista de resultados como uma nova coluna do DataFrame
df['Resultado'] = resultados

# Converter o DataFrame para JSON
json_resultado = df.to_json(orient='records')

# Exibir o JSON resultante
print(json_resultado)
```

```python
import pandas as pd
import gprofiler as gp

# Ler o DataFrame
df = pd.read_csv('caminho/do/arquivo.csv')

# Criar uma lista para armazenar os resultados
resultados = []

# Iterar sobre cada linha do DataFrame
for i, row in df.iterrows():
    # Realizar a query no GProfiler
    res = gp.profile([row['Coluna 1'], row['Coluna 2']], organism='hsapiens', user_threshold=0.05)
    # Verificar se o resultado atende às condições
    if 'source' in res and ('KEGG' in res['source'] or 'BP' in res['source']) and res['intersection_size'] == 2:
        resultados.append(1)
    else:
        resultados.append(0)

# Adicionar a lista de resultados ao DataFrame
df['Resultado'] = resultados

# Exibir o DataFrame final
print(df)
```
