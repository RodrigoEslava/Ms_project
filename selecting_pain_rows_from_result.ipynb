```python
import pandas as pd
```

```python
pain_list = pd.read_csv("4741_pain_id_sym_merged.csv", sep  = " ")
predict_list = pd.read_csv("predicted_edges_genes_less_known_pain_edges (1)", sep = " ")

pain_list
```

```python
predict_list
```

```python
# transforma as listas em conjuntos

conjunto_pain = set(pain_list["NCBI gene (formerly Entrezgene) ID"])
conjunto_predicted = set(predict_list["gene_2"])

conjunto_pain
```

```python
conjunto_predicted
```

```python
# encontra os elementos em comum
intersecao = conjunto_pain & conjunto_predicted
```

```python
intersecao
```

```python
print(len(intersecao))
```

```python
# Selecionar as linhas do dataset predicted que tem genes relacinados a dor na coluna gene_2


selected_pain_pain_rows = predict_list[predict_list["gene_2"].isin(intersecao)]

selected_pain_pain_rows
```

```python
selected_pain_pain_rows.to_csv("full_table_pain_pain_rown.csv", sep = " ", index = None)
```

```python
#Selecionar o top 10
top10_selected_pain_pain_rows = selected_pain_pain_rows[:10]

top10_selected_pain_pain_rows
```

```python
top10_selected_pain_pain_rows.to_csv("top10_table_pain_pain_rown.csv")
```
